```python
from IPython.display import display, HTML, clear_output
import time

# --- CSS HÍBRIDO ---
estilo = """
<style>
    .status-panel { background-color: #1a1a1a; color: #00ff41; padding: 10px; border: 2px solid #333; width: 550px; border-radius: 8px 8px 0 0; font-family: monospace; }
    .terminal-window {
        background-color: #000; color: #fff; padding: 10px; border: 1px solid #333;
        width: 550px; height: 180px; overflow-y: auto; font-family: monospace;
        display: flex; flex-direction: column-reverse;
    }
    .cpu-log { color: #00ff41; border-bottom: 1px dotted #222; padding: 3px 0; font-size: 0.9em; border-left: 3px solid #ffcc00; padding-left: 5px; }
    .counter { color: #ffcc00; font-weight: bold; font-size: 1.2em; }
</style>
"""

processos = []
historico_cpu = []

def atualizar_tela(msg="Aguardando...", tag="SISTEMA"):
    # CORREÇÃO AQUI: Aspas adicionadas corretamente na linha do join
    logs_html = "".join([f"<div class='cpu-log'>[CPU] {log}</div>" for log in reversed(historico_cpu)])

    qtd_processos = len(processos)

    interface = f"""
    <div class='status-panel'>
        <b>[KERNEL MONITOR]</b><br>
        > PROCESSOS EM FILA (RAM): <span class='counter'>{qtd_processos}</span><br>
        > ÚLTIMA AÇÃO: <span style='color:white'>[{tag}] {msg}</span>
    </div>
    <div class='terminal-window'>{logs_html}</div>
    """

    clear_output(wait=True)
    display(HTML(estilo + interface))
    time.sleep(0.1)

# --- EXECUÇÃO ---
atualizar_tela("Boot Finalizado. Contador em 0.")

while True:
    print("\n" + "="*58)
    try:
        cmd = input("COMANDO [1:Add | 2:Rodar | 0:Sair]: ")
    except EOFError:
        break

    if cmd == '1':
        nome = input("NOME DO PROCESSO: ").upper()
        if nome:
            processos.append(nome)
            atualizar_tela(f"Processo {nome} alocado.", "MEMÓRIA")

    elif cmd == '2':
        if processos:
            p = processos.pop(0) # O contador desce aqui automaticamente
            timestamp = time.strftime('%H:%M:%S')
            historico_cpu.append(f"Executado: {p} ({timestamp})")

            atualizar_tela(f"CPU processando: {p}...", "KERNEL")
            time.sleep(0.5)
            atualizar_tela(f"CPU liberada.", "CPU")
        else:
            atualizar_tela("Fila vazia! Contador em 0.", "AVISO")

    elif cmd == '0':
        print("\n[!] Sistema encerrado.")
        break
```
