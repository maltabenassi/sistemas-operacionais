```python
import platform
import os
import psutil # Biblioteca para ler dados do hardware

def auditoria_do_sistema():
    print("=== RELATÓRIO DE INTERAÇÃO COM O S.O. ===")

    # 1. Identificando o "Gigante" (Windows vs Linux vs Unix)
    sistema = platform.system()
    print(f"[NÓ DO KERNEL] Sistema Operacional detectado: {sistema}")
    print(f"[VERSÃO] Release do Software: {platform.release()}")

    # 2. Gerenciamento de Memória (Pilar do seu resumo)
    memoria = psutil.virtual_memory()
    total_gb = memoria.total / (1024**3) # Convertendo bytes para GB
    print(f"[MEMÓRIA RAM] Total gerenciado pelo SO: {total_gb:.2f} GB")

    # 3. Sistema de Arquivos e Diretórios
    diretorio_atual = os.getcwd()
    print(f"[ARQUIVOS] Você está operando no diretório: {diretorio_atual}")

    # 4. Verificando o Processador (Arquitetura)
    print(f"[CPU] Arquitetura da Máquina: {platform.machine()}")

    print("\n--- Auditoria Concluída com Sucesso ---")

auditoria_do_sistema()
```
